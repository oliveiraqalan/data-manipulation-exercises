### Tarefa 1. Importe o dataset e salve os dados em um dataframe

Os dados estão salvos no arquivo ***datasets/users_dataset.txt***.

Salve os dados em uma variável de nome *users*.

````python
import pandas as pd
import numpy as np
dados = pd.read_csv('https://raw.githubusercontent.com/awarischool/data-manipulation-exercises/master/datasets/users_dataset.csv',encoding='utf8',sep='|')
df = pd.DataFrame(dados)
````

### Tarefa 2. Descubra qual a idade média por ocupação.

*Dica: use a função groupby do Pandas na coluna occupation*

````python
# Seu código
#df.age.groupby(df['occupation']).mean()
df[['age']].groupby(df['occupation']).mean()
````

### Tarefa 3. Para cada ocupação descubra a idade mínima e máxima.

````python
# Seu código
# df.age.groupby(df['occupation']).min()  
# df.age.groupby(df['occupation']).max()
#df[['age']].groupby(df['occupation']).min()
#df[['age']].groupby(df['occupation']).max()
df.groupby(['occupation']).agg({'age':np.min})
````

````python
df.groupby(['occupation']).agg({'age':np.max})
````

### Tarefa 4. Para cada combinação de ocupação e sexo, calcule a idade média.

Exemplo:

```
occupation     gender
administrator  F         40.638889
               M         37.162791
artist         F         30.307692
               M         32.333333
```

````python
# Seu código
df.groupby(['occupation','gender']).agg({'age':np.mean})
#df.groupby(['occupation','gender']).age.mean()
````

### Tarefa 5. Para cada ocupação calcule a porcentagem de homens e mulheres.

Exemplo:

```
occupation     gender
administrator  F          45.569620
               M          54.430380
artist         F          46.428571
               M          53.571429
doctor         M         100.000000
educator       F          27.368421
               M          72.631579
```

````python
# Seu código
grouped_df = df.groupby(['occupation','gender']).agg({'gender':len})
percents_df = grouped_df.groupby(level=0).apply(lambda x: 100 * x / float(x.sum()))
percents_df['gender']

````
